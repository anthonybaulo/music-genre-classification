```python
import os
import numpy as np
import librosa
import librosa.display
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
META_DIR = Path('data/fma_metadata')
AUDIO_DIR = Path('data/fma_small')
CONVERTED_DIR = Path('data/converted/rock_inst')
```

# Load and check

```python
npzfile = np.load(CONVERTED_DIR/'train_arr.npz')
X_train = npzfile['X']
y_train = npzfile['y']
print(X_train.shape, y_train.shape)
```

```python
librosa.display.specshow(X_train[0].T, y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
```

```python
X_train[0][0].max()
```

```python
X_raw = librosa.core.db_to_power(X_train[0], ref=1.0)
X_log = np.log(X_raw)

librosa.display.specshow(X_log.T, y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
```
